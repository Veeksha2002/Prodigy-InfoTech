```python
df['A']=df[['Annual Income (k$)']]
df['B']=df[['Spending Score (1-100)']]
```

```python
x=df[['A','B']]
x.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(x['A'],x['B'],s=30,c='b')
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
Kmean=KMeans(n_clusters=5)
Kmean.fit(x)
centers=Kmean.cluster_centers_
print(Kmean.cluster_centers_)
```

```python
clusters=Kmean.fit_predict(x)
df['label']=clusters
df.head(10)
```

```python
col=['blue','purple','maroon','black','salmon']
for i in range(5):
    a=col[i]
    plt.scatter(df.A[df.label==i],df.B[df.label==i],c=a,label='Cluster')
plt.scatter(centers[:,0],centers[:,1],marker='*',s=300,c='r',label='Centroid')
```

```python
y=x.loc[:,['A','B']].values
y
```

```python
wcss=[]
for j in range(1,11):
    kmeans=KMeans(n_clusters=j,init='k-means++')
    kmeans.fit(y)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2,color='indigo',marker='8')
plt.xlabel('K Value')
plt.ylabel('WCSS')
plt.show()
```
